```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
temp_max=np.array([ 39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25])
temp_min=np.array([ 21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18])
n=np.linspace(1,12,num=12,endpoint=True,dtype=int)
```

```python
n
```

```python
cum_max=np.cumsum(temp_max)

cum_min=np.cumsum(temp_min)
```

```python
def average(n,x):
    return np.divide(x,n)
```

```python
average(cum_max,n)
```

```python
days = np.linspace(1, 12, num=12)

plt.figure()
plt.plot(n, temp_max, 'ro')
plt.plot(days, average(days,cum_max), 'r-')
plt.plot(n, temp_min, 'bo')
plt.plot(days, average(days,cum_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')

plt.show()
```
